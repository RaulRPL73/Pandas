1. Autos con Millaje, precio y modelo

```python
df_milla_sorted = df.sort_values(by=['Mileage', 'Price' ,'Model'], ascending=[True, True, True])
df_milla_sorted .head(10)
```

2. Lista de modelo conforme a su año

```python
df_modelo_sorted = df.sort_values(by=['Model', 'Year'], ascending=[True, True])
df_modelo_sorted .head(10)
```

3. Grafica de los tipos de carros que los usuarios utilizan

```python
df_tipo = df[['Transmission']].value_counts().reset_index(name='Count')

colors = sns.color_palette('pastel')

plt.figure(figsize=(8, 8))
plt.pie(df_tipo['Count'], labels=df_tipo['Transmission'], autopct='%1.1f%%', startangle=140, colors=colors)
plt.title('Distribución de Tipos de Transmisión')
plt.show()
```

4. Grafica de los tipos de combustible que se utiliza

```python
df_combustible = df[['Fuel_Type']].value_counts().reset_index(name='Count')

colors = sns.color_palette('pastel')

plt.figure(figsize=(8, 8))
plt.pie(df_combustible['Count'], labels=df_combustible['Fuel_Type'], autopct='%1.1f%%', startangle=140, colors=colors)
plt.title('Distribución de Tipos de combustible')
plt.show()
```

5. Grafica de los tipos de combustible que se utiliza

```python
df_doors = df[['Doors']].value_counts().reset_index(name='Count')

colors = sns.color_palette('pastel')
labels = [f"Doors: {door}" for door in df_doors['Doors']]

plt.figure(figsize=(8, 8))
plt.pie(df_doors['Count'], labels=labels, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title('Distribución de Tipos de puertas')
plt.show()
```
