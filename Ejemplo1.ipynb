```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Spotify_Youtube.csv")
```

```python
df.head()
```

1. Obtener data frame de youtube
2. Ordenarlos conforme la cantidad de like descendente
3. Graficar los primeros 10 artistas conforme a la cantidad de likes
4. Repetir proceso con spotify
5. Graficar a los mejores artistas conforme sus likes 
6. Compararlo con spotify

1

```python
df_youtube = df[["Artist", "Url_youtube", "Title", "Likes"]]
df_youtube.head()
```

2

```python
df_youtube_sorted = df_youtube.sort_values(by="Likes", ascending=False)
df_youtube_sorted.head()
```

```python
top_visits = df_youtube_sorted.head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x='Likes', y='Title', data=top_visits, palette='viridis')

plt.title("Top 10 canciones por likes")
plt.xlabel('Likes')
plt.ylabel('Title')
plt.show()
```

Suma de likes de artistas

```python
top_visits_sum = df_youtube_sorted.groupby('Artist').sum().sort_values(by='Likes', ascending=False).head(10)
top_visits_sum.head(10)
```

```python


plt.figure(figsize=(10, 6))
sns.barplot(x='Likes', y='Artist', data=top_visits_sum, palette='viridis')

plt.title("Top 10 canciones por likes")
plt.xlabel('Likes')
plt.ylabel('Artist')
plt.show()
```

```python
df_spotify = df[["Artist", "Url_spotify", "Track", "Energy"]]
df_spotify.head()
```

```python
df_spotify_sorted = df_spotify.sort_values(by="Energy", ascending=False)
df_spotify_sorted.head()
```

```python
print(df_spotify_sorted.columns)
```

```python
top_tracks = df_spotify_sorted.head(10)
plt.figure(figsize=(10,6))
sns.barplot(x='Energy', y='Track', data=top_tracks, palette='viridis')

plt.title('Top 10 canciones por energia')
plt.xlabel('Energy')
plt.ylabel('Track')
plt.show()
```

Suma de Energia de artistas

```python
top_tracks_sum = df_spotify_sorted.groupby('Artist').sum().sort_values(by='Energy', ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x='Energy', y='Artist', data=top_tracks_sum, palette='viridis')

plt.title("Top 10 canciones por Energia")
plt.xlabel('Likes')
plt.ylabel('Artist')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Energy', y='Artist', data=top_tracks_sum, palette='viridis')

plt.title("Top 10 canciones por Energia")
plt.xlabel('Likes')
plt.ylabel('Artist')
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(x='Likes', y='Artist', data=top_visits_sum, palette='viridis')

plt.title("Top 10 canciones por likes")
plt.xlabel('Likes')
plt.ylabel('Artist')
plt.show()
```

Unir ambos en un solo DataFrame

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 6))
```

Primer grÃ¡fico (Spotify - EnergÃ­a)

```python
colors = sns.color_palette("coolwarm", len(top_tracks_sum))
sizes = top_tracks_sum["Energy"] * 100  # Ajustamos el tamaÃ±o de los cÃ­rculos

sns.barplot(
    x=top_tracks_sum.index, y=top_tracks_sum["Energy"], palette=colors, alpha=0.7, ax=axes[0]
)

for i, energy in enumerate(top_tracks_sum["Energy"]):
    axes[0].scatter(i, energy, s=sizes[i], color=colors[i], edgecolors="black", alpha=0.8)

axes[0].set_ylabel("Nivel de EnergÃ­a")
axes[0].set_xlabel("Artista")
axes[0].set_title("Top 10 Artistas con Mayor EnergÃ­a en Spotify")
axes[0].set_xticklabels(top_tracks_sum.index, rotation=45, ha="right")
```

Segundo grÃ¡fico (YouTube - Likes)

```python
sns.barplot(
    x=top_visits_sum.index, y=top_visits_sum["Likes"], palette="rocket", ax=axes[1]
)

axes[1].set_ylabel("NÃºmero de Likes")
axes[1].set_xlabel("Artista")
axes[1].set_title("Top 10 Artistas con MÃ¡s Likes en YouTube")
axes[1].set_xticklabels(top_visits_sum.index, rotation=45, ha="right")
```
